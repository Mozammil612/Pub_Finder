```python
#id, name, address, postcode, easting, northing, latitute, longitute, local_authority
df = pd.DataFrame(
    np.row_stack([df.columns, df.values]),
    columns=['fsa_id', 'pub', 'address', 'postcode', 'east', 'north', 'latitute', 'longitute', 'local_authority']
)
```

```python
df.head()
```

```python
df.info()
```

```python
#Converting Data Types
df['fsa_id'] = df['fsa_id'].astype('int')
df['pub'] = df['pub'].astype('string')
df['address'] = df['address'].astype('string')
df['postcode'] = df['postcode'].astype('string')
df['east'] = df['east'].astype('int')
df['north'] = df['north'].astype('int')
df['local_authority'] = df['local_authority'].astype('string')
```

```python
print(df['latitute'])
```

```python
df['latitute'] = pd.to_numeric(df['latitute'], errors='coerce')
df['longitute'] = pd.to_numeric(df['longitute'], errors = 'coerce')
```

```python
print(df['latitute'])
```

```python
print(df['longitute'])
```

```python
## Replacing missing values from "latitude" and "longitude" column with their median values.
df["latitute"].replace(np.nan,df["latitute"].median(),inplace=True)
df["longitute"].replace(np.nan,df["longitute"].median(),inplace=True)
```
